```python
from random import randint
import numpy as np
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels=[]
train_samples=[]
```

```python
for i in range(50):
    ry=randint(13,64)
    train_samples.append(ry)
    train_labels.append(1)
    ro=randint(65,100)
    train_samples.append(ro)
    train_labels.append(0)
```

```python
for i in range(1000):
    ry=randint(13,64)
    train_samples.append(ry)
    train_labels.append(0)
    ro=randint(65,100)
    train_samples.append(ro)
    train_labels.append(1)
```

```python
for i in train_samples:
    print(i)
for i in train_labels:
    print(i)
```

```python
train_samples=np.array(train_samples)
train_labels=np.array(train_labels)
train_samples,train_labels=shuffle(train_samples,train_labels)
scaler=MinMaxScaler(feature_range=(0,1))
scaler_train_samples=scaler.fit_transform((train_samples.reshape(-1,1)))
for i in scaler_train_samples:
    print(i)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation,Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
model=Sequential([
    Dense(units=16, input_shape=(1,),activation="relu"),
    Dense(units=32,activation="relu"),
    Dense(units=2,activation="softmax")
                 ])
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001),loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(x=scaler_train_samples,y=train_labels,batch_size=10,epochs=30,verbose=2)
```

```python
model.fit(x=scaler_train_samples,y=train_labels,validation_split=0.1,batch_size=10,epochs=30,shuffle=True,verbose=2)
```

```python
test_samples=[]
test_labels=[]
```

```python
for i in range(10):
    ry=randint(13,64)
    test_samples.append(ry)
    test_labels.append(1)
    ro=randint(65,100)
    test_samples.append(ro)
    test_labels.append(0)
```

```python
for i in range(200):
    ry=randint(13,64)
    test_samples.append(ry)
    test_labels.append(0)
    ro=randint(65,100)
    test_samples.append(ro)
    test_labels.append(1)
```

```python
test_samples=np.array(test_samples)
test_labels=np.array(test_labels)
test_samples,test_labels=shuffle(test_samples,test_labels)
scaler_test_samples=scaler.fit_transform(test_samples.reshape(-1,1))
```

```python
prediction=model.predict(x=scaler_test_samples,batch_size=10,verbose=0)
```

```python
rp=[]
for i in prediction:
    rp.append(np.round(i[1]))
```

```python
from sklearn.metrics import confusion_matrix
%matplotlib inline
import itertools
import matplotlib.pyplot as plt
cm=confusion_matrix(y_true=test_labels,y_pred=rp)
```
